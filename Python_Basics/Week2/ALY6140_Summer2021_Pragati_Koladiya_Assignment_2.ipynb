```python
import pandas as pd # to create dataframe for accessing CSV and Excel
import pyabf # for ABF file
import numpy as np
import matplotlib.image as mpimg # for image file 
import matplotlib.pyplot as plt # to plot the image

#to ignore the future warnings 
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## Used functions to open the file and print header

```python
def get_file_open(F1,F2,F3,F4):
    
    print("Binary File Path: {}".format(F1))
    print("----------------------------------------------Binary head-------------------------------------------")
    Binary_File = pyabf.ABF(F1,'r')
    binary1 = print(Binary_File)
    print("-------------------------------------------Binary header text---------------------------------------")
    binary = print(Binary_File.headerText)

    print("====================================================================================================")
    print("CSV File Path: {}".format(F2))
    CSV_File = pd.read_csv(F2,'r')
    csv = print(CSV_File.head())
    
    print("====================================================================================================")
    print("Excel File Path: {}".format(F3))
    Excel_File = pd.read_excel(F3)
    excel = print(Excel_File.head())

    print("====================================================================================================")
    print("Image File Path: {}".format(F4))
    Image_File = mpimg.imread(F4)
    img = plt.imshow(Image_File)
    
    return binary1, binary, csv, excel, img

if __name__ == "__main__":

    F1 = '/Users/pragati/OneDrive - Northeastern University/Q6-Summer/Week2/16o03002.abf'
    F2 = '/Users/pragati/OneDrive - Northeastern University/Q6-Summer/Week2/Nuclei.csv'
    F3 = '/Users/pragati/OneDrive - Northeastern University/Q6-Summer/Week2/Financial Sample.xlsx'
    F4 = '/Users/pragati/OneDrive - Northeastern University/Q6-Summer/Week2/Nuclei_Msk.jpg'
    get_file_open(F1,F2,F3,F4)
    
```

<i>Used <b>Main function</b> to call the user defined function by passing all four file paths</i>

- The user define functions is being used for opening the file and to print the head of each file.
    - Binary File
     - Binary files are stored with .ABF extension so to open the binary file we need to import the pyabf package.
     - Binary files object has no attribute 'head' so for that if we want to print the head we can either print the object itself or can use headerText as shown in the above function.
     
    - CSV File
     - For the csv file we need to use the pandas object to use read_csv().
     - To print the head pandas object does support head(). This will print top 5 records from the file.
     
    - Excel File
     - For excel file as well we need to use the pandas obejct. By callig read_excel() it will load the file into dataframe.
     - To view the top few records we have use head() as well.
     
    - Image File 
     - To open image file, need to import the matplotlib.image
     - To view the image, plt.imshow() is being used.
